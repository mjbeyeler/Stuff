```python
# Original author: Daniel Krefl
# modified by: mjbeyeler

import numpy as np
from scipy.stats import ncx2
from scipy.stats import chi2
import matplotlib.pyplot as plt
```

```python
def t(sig):
    t = chi2.ppf(1-sig,1)
    return t
```

```python
def P(n,MAF,beta):
    q2 = 2*MAF*(1-MAF)*beta**2
    nc = n*q2/(1-q2)
    return 1-ncx2.cdf(t(5*10**-4),1,nc)

maf01 = []
pow01 = []
for i in np.linspace(0.001,0.5,1000):
    maf01.append(i)
    pow01.append(P(1000,i,0.01))

maf05 = []
pow05 = []
for i in np.linspace(0.001,0.5,1000):
    maf05.append(i)
    pow05.append(P(1000,i,0.05))

maf1 = []
pow1 = []
for i in np.linspace(0.001,0.5,1000):
    maf1.append(i)
    pow1.append(P(1000,i,0.1))

maf15 = []
pow15 = []
for i in np.linspace(0.001,0.5,1000):
    maf15.append(i)
    pow15.append(P(1000,i,0.15))



def P(n,MAF,beta):
    q2 = 2*MAF*(1-MAF)*beta**2
    nc = n*q2/(1-q2)
    return 1-ncx2.cdf(t(2.5*10**-4),1,nc)

maf01o = []
pow01o = []
for i in np.linspace(0.001,0.5,1000):
    maf01o.append(i)
    pow01o.append(P(1000,i,0.01))

maf05o = []
pow05o = []
for i in np.linspace(0.001,0.5,1000):
    maf05o.append(i)
    pow05o.append(P(1000,i,0.05))

maf1o = []
pow1o = []
for i in np.linspace(0.001,0.5,1000):
    maf1o.append(i)
    pow1o.append(P(1000,i,0.1))

maf15o = []
pow15o = []
for i in np.linspace(0.001,0.5,1000):
    maf15o.append(i)
    pow15o.append(P(1000,i,0.15))
```
